# 折線圖

適用：會隨時間變動的值，呈現趨勢變化 <br/>
實際應用：匯率 <br/>
由匯率折線圖可看出，美金從九月中旬至今仍持續貶值中

# 折線圖 in pandas plot

折線圖可以簡單的在 pandas 資料使用 .plot() 就可以畫出來 <br/>
其中 x 軸部分為資料集的 index，y 軸為資料集欄位值，如下圖紅框資料

```python
import pandas as pd
import numpy as np
import random
```

```python
pd.date_range('1/1/2020', periods=200)
```

```python
#單一折線圖
ts = pd.Series(np.random.randn(200), index=pd.date_range('1/1/2020', periods=200))
ts = ts.cumsum()   #cumsum()累加量
print(ts)
ts.plot()
```

如果資料有多個欄位時，也可以同時話出多筆資料在同一張圖上。

```python
#多個折線圖同時畫出
df = pd.DataFrame(np.random.randn(200, 4), index=pd.date_range('1/1/2020', periods=200), columns=["A","B","C","D"])
df = df.cumsum()
print(df)
df.plot();
```

# 圓餅圖 in pandas plot

圓餅圖，其中資料 index 為圓餅圖類別 ABCD，欄位部份為 ABCD 所對應數值， .plot.pie() 會依據所對的數值計算比例畫出圓餅圖。<br/>
參數 name 為圓餅圖的名稱

```python
#圓餅圖
df = pd.Series(np.random.rand(4), index=["A","B","C","D"],name="title")
print(df)
df.plot.pie()
```

# 長條圖

適用：不同種類資料，在不同時間點的變化 <br/>
實際應用：人口成長變化 <br/>
由人口長條圖可看出，從 102 年到 107 年新北市人口都是六都之冠；桃園市是六都中人口成長最快速的城市

# 長條圖 in pandas plot

長條圖，其中資料 index 為 x 軸，欄位部份為 AB 所對應數值，.plot.bar() 會依據所對的數值畫出長條圖。 <br/>
參數 stacked 可以將長條圖疊加

```python
#長條圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.bar();
```

參數 stacked 可以將長條圖疊加


```python
df.plot.bar(stacked=True);
```

# 箱型圖

適用：完整呈現數值分布的統計圖表 <br/>
實際應用：薪資水平 <br/>
由薪資箱型圖可看出，廣州的最低薪資是四個城市中最高的，而深圳是薪資水平最高的城市


# 箱型圖 in pandas plot

箱型圖，其中資料 index 為 x 軸，欄位部份為 AB 所對應數值，.boxplot() 會依據所對的數值畫出箱型圖。

```python
#箱型圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.boxplot()
```

# 散佈圖

適用：呈現相關數值間的關係<br/>
實際應用：性別與體重的關係<br/>
由散佈圖可看出，大多數的人，身高與體重成正比關係，而男性的身高體重趨勢，都高於女性

# 散佈圖 in panda plot

散佈圖，將資料欄位以 x 軸為 A，y 軸為 B 畫出散佈圖

```python
#散佈圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.scatter(x='A', y='B')
```
