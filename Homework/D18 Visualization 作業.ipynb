# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

# 每個點的颜色和該點的X值+Y值的和相关
color = X + Y

# 使用colormap
cmap = plt.get_cmap('coolwarm')

plt.scatter(X, Y, s=75, alpha=0.5, c=color, cmap=cmap)

'''
#作業

#方法不只有一種

'''

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

```python
# 使用不同colormap 'plasma'
cmap = plt.get_cmap('plasma')
plt.scatter(X, Y, c=color, cmap=cmap)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

```python
# 使用不同colormap 'Oranges'
cmap = plt.get_cmap('Oranges')
plt.scatter(X, Y, c=color, cmap=cmap)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

```python
# 使用不同colormap 'summer'
cmap = plt.get_cmap('summer')
plt.scatter(X, Y, c=color, cmap=cmap)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

```python
# 使用不同colormap 'rainbow'
cmap = plt.get_cmap('rainbow')
plt.scatter(X, Y, c=color, cmap=cmap)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
